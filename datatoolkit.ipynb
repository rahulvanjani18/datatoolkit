
# 1. What is NumPy, and why is it widely used in Python?
> NumPy is a Python library used for numerical computing. It provides efficient array operations, mathematical functions, and tools for linear algebra, making it ideal for scientific and data-based applications.
# ---------------------------------------------------------------------

# 2. How does broadcasting work in NumPy?
> Broadcasting allows NumPy to perform arithmetic operations on arrays of different shapes by automatically expanding them to a common shape.
# ---------------------------------------------------------------------

# 3. What is a Pandas DataFrame?
> A DataFrame is a 2-dimensional, labeled data structure in Pandas similar to a table or spreadsheet, used for handling and analyzing structured data.
# ---------------------------------------------------------------------

# 4. Explain the use of the groupby() method in Pandas
> The groupby() method is used to group data by one or more columns and perform aggregate functions like sum, mean, or count on each group.
# ---------------------------------------------------------------------

# 5. Why is Seaborn preferred for statistical visualizations?
> Seaborn simplifies the creation of attractive and informative statistical plots with built-in support for complex plots like boxplots, violin plots, and heatmaps using fewer lines of code.
# ---------------------------------------------------------------------

# 6. What are the differences between NumPy arrays and Python lists?
> NumPy arrays are more memory-efficient, faster, and support vectorized operations, while Python lists are more flexible but slower and lack mathematical capabilities.
# ---------------------------------------------------------------------

# 7. What is a heatmap, and when should it be used?
> A heatmap is a graphical representation of data where individual values are shown as colors. It's used to visualize the magnitude of values across a matrix or correlation between variables.
# ---------------------------------------------------------------------

# 8. What does the term “vectorized operation” mean in NumPy?
> Vectorized operations refer to performing operations on entire arrays without using explicit loops, making the code faster and more efficient.
# ---------------------------------------------------------------------

# 9. How does Matplotlib differ from Plotly?
> Matplotlib creates static plots and is better for simple custom plots, while Plotly supports interactive, web-based visualizations with zoom and hover capabilities.
# ---------------------------------------------------------------------

# 10. What is the significance of hierarchical indexing in Pandas?
> Hierarchical indexing allows multiple levels of indexing in a DataFrame or Series, enabling more complex data representation and efficient data selection.
# ---------------------------------------------------------------------

# 11. What is the role of Seaborn’s pairplot() function?
> pairplot() creates a matrix of scatter plots for visualizing pairwise relationships in a dataset, often used for exploratory data analysis.

# 12. What is the purpose of the describe() function in Pandas?
> describe() generates summary statistics such as count, mean, std, min, and max for numeric columns in a DataFrame.

# 13. Why is handling missing data important in Pandas?

# 14. What are the benefits of using Plotly for data visualization?

# 17. Explain the difference between apply() and map() in Pandas

```python
#2 Write a Pandas script to find the mean of a specific column in a DataFrameA
a = df['Sales'].mean()
print(a)
```

```python
#3Create a scatter plot using Matplotlib
plt.figure(figsize=(8,5))
plt.scatter(df['QUANTITYORDERED'], df['Sales'], color='orange', alpha=0.6)
plt.title('Quantity Ordered vs Sales')
plt.xlabel('Quantity Ordered')
plt.ylabel('Sales')
plt.show()
```

```python
#4 How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap
df1 = pd.DataFrame({
    'A': [1, 2, 3, 4, 5],
    'B': [5, 4, 3, 2, 1],
    'C': [2, 3, 4, 5, 6]
})


corr_matrix = df1.corr()


plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix Heatmap")
plt.show()
```

```python
#5Generate a bar plot using Plotly
plt.figure(figsize=(8,5))
plt.bar(df['Sales'].index, df['Sales'].values, color='green')
plt.title(' Sales ')
plt.xlabel('sales')
plt.ylabel(' Sales')
plt.xticks(rotation=45)
plt.show()
```

```python
#6 Create a DataFrame and add a new column based on an existing column
avg_sales = df['Sales']/df['QUANTITYORDERED']
df['avg_sales'] = avg_sales
df
```

```python
#7 Write a program to perform element-wise multiplication of two NumPy arrays.
np1 = np.array([1,2,3])
np2 = np.array([4,5,6])
np3 = np1*np2
print(np3)
```

```python
#8Create a line plot with multiple lines using Matplotlib
sales_over_time = df.groupby('ORDERDATETIME')['Sales'].sum()
plt.figure(figsize=(10,5))
plt.plot(sales_over_time.index, sales_over_time.values, marker='o', linestyle='-', color='purple')
plt.title('Total Sales Over Time')
plt.xlabel('Order Date')
plt.ylabel('Total Sales')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#9 Generate a Pandas DataFrame and filter rows where a column value is greater than a threshold
df[df['Sales'] > 3000]
```

```python
#10 Create a histogram using Seaborn to visualize a distribution
plt.figure(figsize=(8,5))
sns.histplot(df['QUANTITYORDERED'], bins=20, kde=True, color='skyblue')
plt.title('Histogram of Quantity Ordered')
plt.xlabel('Quantity Ordered')
plt.ylabel('Frequency')
plt.show()
```
